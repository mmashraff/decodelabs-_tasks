# Outlier Detection

```python
for col in ['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers")
```

# Visualization

```python
df['Month'] = df['Date'].dt.to_period('M')
monthly = df.groupby('Month')['TotalPrice'].sum()

monthly.plot(xlabel='Month', ylabel='Revenue', color='steelblue')
plt.title('Monthly Revenue Trend')
```

```python
Revenue_product = df.groupby('Product')['TotalPrice'].sum()

Revenue_product.sort_values(ascending = True).plot.bar(figsize = (8, 4), color = 'teal', title = 'Revenue by product (total $)')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
df['OrderStatus'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Order Status Distribution')
plt.ylabel('')
```

```python
ReferralSource_Revenue = df.groupby('ReferralSource')['TotalPrice'].sum()

ReferralSource_Revenue.sort_values(ascending = True).plot.barh(figsize = (8, 4), color = 'purple', title = 'Referral source vs revenue')
plt.xlabel('Revenue')
plt.ylabel('ReferralSource')
plt.tight_layout()
plt.show()
```
